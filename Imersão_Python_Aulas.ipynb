```python
import pandas as pd
import plotly.express as px

```

```python
def_principal = pd.read_excel("/content/tabela_acoes.xlsx",sheet_name="Principal")
```

```python
def_principal.head(10)
```

```python
def_total_acoes = pd.read_excel("/content/tabela_acoes.xlsx",sheet_name="Total_de_acoes")
def_total_acoes.head(10)
```

```python
def_ticker = pd.read_excel("/content/tabela_acoes.xlsx",sheet_name="Ticker")
def_ticker.head(10)
```

```python
def_chatgpt = pd.read_excel("/content/tabela_acoes.xlsx",sheet_name="Chatgpt")
def_chatgpt.head(10)
```

```python
def_principal=def_principal[['Ativo','Data','Último (R$)', 'Var. Dia (%)']].copy()
```

```python
def_principal.head(10)
```

```python
def_principal= def_principal.rename(columns={'Último (R$)': 'valor_final', 'Var. Dia (%)':'var_dia_pct'}).copy()
```

funçao copy() faz com que seja criado e substituido o dataframe

```python
def_principal.head(10)
```

```python
def_principal['var_pct']= def_principal['var_dia_pct']/100
```

```python
def_principal['valor_inicial']= def_principal['valor_final']/(def_principal['var_pct']+1)
```

```python
def_principal.head(10)
```

```python
def_principal=def_principal.merge(def_total_acoes,left_on='Ativo', right_on='Código',how='left')
```

```python
def_principal.head(10)
```

```python
def_principal= def_principal.drop(columns=['Código'])
```

```python
def_principal.head(10)
```

```python
def_principal['variacao_rs'] = (def_principal['valor_final']- def_principal['valor_inicial'])*def_principal['Qtde. Teórica']
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
def_principal.head(10)
```

```python
def_principal['Qtde. Teórica'
] = def_principal['Qtde. Teórica'
].astype(int)
```

```python
def_principal.head(10)
```

```python
def_principal= def_principal.rename(columns={'Qtde. Teórica': 'qtd_teorica'}).copy()
```

```python
def_principal
```

```python
def_principal['resultado']= def_principal['variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else('Desceu' if x < 0 else 'Estavel'))
```

```python
def_principal
```

```python
def_principal= def_principal.merge(def_ticker, left_on='Ativo',right_on='Ticker', how='left')
def_principal= def_principal.drop(columns=['Ticker'])
```

```python
def_principal
```

```python
def_principal= def_principal.merge(def_chatgpt, left_on='Nome',right_on='Nome da empresa', how='left')
def_principal= def_principal.drop(columns=['Nome da empresa'])
```

```python
def_principal
```

```python
def_principal['cat_idade']= def_principal['Idade (anos)'].apply(lambda x : 'Mais de 100' if x > 100 else ('Menos de 50' if x < 50 else 'Entre 50 e 100'))
def_principal
```

```python
maior = def_principal['variacao_rs'].max()
menor = def_principal['variacao_rs'].min()
media = def_principal['variacao_rs'].mean()
media_subiu = def_principal[def_principal['resultado']=='Subiu']['variacao_rs'].mean()
media_desceu = def_principal[def_principal['resultado']=='Desceu']['variacao_rs'].mean()
```

```python
def_principal_subiu =  def_principal[def_principal['resultado']=='Subiu']
```

reset_index é pra continuar sendo um dataframe

```python
df_analise_segmento = def_principal_subiu.groupby('Segmento')['variacao_rs'].sum().reset_index()
df_analise_segmento
```

```python
df_analise_saldo = def_principal.groupby('resultado')['variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo,x='resultado', y= 'variacao_rs', text='variacao_rs', title='Variacao reais por Resultado')
fig.show()
```
